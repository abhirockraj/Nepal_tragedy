```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= pd.read_csv('train_values.csv')
x2= pd.read_csv('train_labels.csv')
```

```python
y_test = x2.drop('building_id', axis=1)
```

```python
y_test.head()
```

```python
x_test=x
x_test.head()
```

```python
x_test.info
```

```python
x_test.describe()
```

```python
x_test = x_test.replace('?', np.nan)
```

```python
x_test.dtypes
```

```python
from sklearn.preprocessing  import LabelEncoder
```

```python
#getting the list of categorical variable
s= (x_test.dtypes == 'object')
obj_cols = list(s[s].index)
```

```python
x=x_test.copy()
l_e = LabelEncoder()
for c in obj_cols:
    x[c]= l_e.fit_transform(x_test[c])
```

```python
x.dtypes
```

```python
import seaborn as sns
for col in x.columns:
    sns.distplot(a=x[col],kde=False)
    plt.show()
```

```python
from xgboost import XGBRegressor
model= XGBRegressor()
model.fit(x , y_test)
```

```python
x_val = pd.read_csv('test_values(1).csv')
```

```python
x_test_f=x_val.copy()
l_e = LabelEncoder()
for c in obj_cols:
    x_test_f[c]= l_e.fit_transform(x_val[c])
```

```python
y_pred = model.predict(x_test_f)
```

```python
y_pred =[int(x) for x in y_pred]
```

```python
print(y_pred)
```

```python
df = pd.DataFrame(data=y_pred, columns=['damage_grad'])
```

```python
df= pd.concat([x_val['building_id'], df], axis=1)
df.head()
```

```python
df.to_csv('output_final.csv', index= False)
```
